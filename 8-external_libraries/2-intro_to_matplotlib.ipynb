```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.arange(10, dtype=np.int64)
y = np.random.rand(10) * 10
print(x, y)
```

## Scatter Plot

```python
plt.scatter(x, y)
plt.show()
```

## Line plot

```python
plt.plot(x, y)
```

```python
y = np.floor(y)
```

## Bar plot

```python
categories = ["male", "female"]
plt.bar(x, y)
plt.show()
```

## Histogram

```python
y = [1, 1, 1, 2, 3, 4, 4]
plt.hist(y)
plt.show()
```

## Pie Chart

```python
categories = ["short", "medium", "tall"]
percentages = [.3, 0.5, 0.2]

plt.pie(percentages, labels=categories)
plt.show()
```
